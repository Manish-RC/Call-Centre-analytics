# MODEL FITTING
Models Applying:<br>
1. Decision Tree
2. Naive Bayes Classifier
3. SVM

```python
from sklearn.model_selection import train_test_split
X = df[['u_city', 'u_occupation', 'u_income_lpa', 'u_religion', 'u_caste',
       'u_sibling', 'u_horoscope', 'website_visit', 'u_age', 'agent_ID',
       'a_city', 'a_horoscopes', 'a_sales_performance', 'Communication Skills',
       'Empathy Level', 'Sales Targets', 'Conversion Rates',
       'If_agent_male', 'Product_user', 'Experienced_agent',
       'Supportive_customer', 'Formal_approach', 'If_user_male', 'u_education',
       'u_family_background', 'A_education']].values
y = df[['label']].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print('Decision Tree Accuracy:', accuracy_dt)

# Naive Bayes classifier
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print('Naive Bayes Accuracy:', accuracy_nb)

# SVM classifier
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print('SVM Accuracy:', accuracy_svm)
```
